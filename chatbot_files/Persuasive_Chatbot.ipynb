```python
# Cell 2 - Step 2


domain_yml = """
intents:
 - greet
 - goodbye
 - user_response
 - user_certain
 - user_buy
 - user_uncertain
 
actions:
 - utter_greet
 - utter_question
 - utter_amplify_pain
 - utter_judge_certainty
 - utter_close
 - utter_loop
 - utter_congrats
 - utter_goodbye
 
templates:
  utter_greet:
  - text: "Hey! How are you?"
  
  utter_question:
  - text: "So what are you looking for today?"
  
  utter_amplify_pain:
  - text: "How many hours can you save by taking a Lime?"
  
  utter_judge_certainty:
  - text: "So how do you feel about this product?"
  
  utter_close:
  - text: "That's wonderful? So how many purchases are we looking at?"
  
  utter_loop:
  - text: "I feel what you're saying. But what do you think of the idea?"
  
  utter_congrats:
  - text: "Congrats! You've made a wise decision."
  
  utter_goodbye:
  - text: "Bye!"
"""

%store domain_yml > domain.yml
print("Done!")
```
